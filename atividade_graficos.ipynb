```python
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv('ecommerce_preparados.csv')
print(df.head().to_string())

plt.figure(figsize=(10, 8))
plt.hist(df['Nota'], bins=25, color='red', alpha=0.5)
plt.title('Histograma de Nota')
plt.xlabel('Nota')
plt.xticks(ticks=range(1, int(df['Nota'].max())))
plt.ylabel('Desconto')
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

df = pd.read_csv('ecommerce_preparados.csv')
print(df.head().to_string())

sns.jointplot(x='Desconto', y='Nota', data=df, kind='scatter')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

df = pd.read_csv('ecommerce_preparados.csv')
print(df.head().to_string())

corr = df[['Nota', 'Preço']].corr()
plt.subplot(2, 2, 3)
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlação Nota Preço')

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv('ecommerce_preparados.csv')
print(df.head().to_string())

plt.figure(figsize=(10, 6))
df['Nota'].value_counts().plot(kind='bar', color='#90ee70')
plt.title('Nota por Avaliação')
plt.xlabel('Nota')
plt.ylabel('Avaliação')
plt.xticks(rotation=0)
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv('ecommerce_preparados.csv')
print(df.head().to_string())

labels = ['Marcas', 'Desconto', 'Material', 'Gênero']
sizes = [15, 30, 45, 10]

plt.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.title("Gráfico de Pizza")
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

df = pd.read_csv('ecommerce_preparados.csv')
print(df.head().to_string())

precos = np.linspace(10, 100, 20)
notas = (precos - 55)**2 / 300 + 1

plt.plot(precos, notas)
plt.title("Notas de Avaliação vs Preço")
plt.xlabel("Preço da Roupa")
plt.ylabel("Nota de Avaliação")
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

df = pd.read_csv('ecommerce_preparados.csv')
print(df.head().to_string())

sns.regplot(x='Nota', y='Desconto', data=df, color='#278f65', scatter_kws={'alpha': 0.5, 'color': '#34c289'})
plt.title('Regressão de Nota por Desconto')
plt.xlabel('Nota')
plt.ylabel('Desconto')
plt.show()
```
